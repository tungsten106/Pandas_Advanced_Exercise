### 11- 高亮缺失值

很明显，虽然上一题找到了全部缺失值所在的行，但是看的很头疼

-> 现在将缺失值进行高亮进一步查看

指路：<font color = '#E36C07'>**2-15**</font>

```python
import math

df_rmna = df[df.isna().any(axis=1)]

def na_color_map(v):
    if type(v) == str:
        return None
    if np.isnan(v):
        color = '#E36C07'
    else:
        return None
    return f"color: {color}"

df_rmna.style.applymap(na_color_map)
```

### 14 - 缺失值补全｜向上填充

从上一小节的查看数据中，不难发现整理数据是按照评分进行降序排列的

因此对于评分列的缺失值处理，我们可以用上一个电影的评分进行填充

-> 现在将评分列的缺失值，替换为上一个电影的评分

```python
na_idx_score = list(df[df["评分"].isna()].index)
na_idx_score
```

```python
df["评分"].fillna(method="ffill", inplace=True)
# sum(df["评分"].isna())  # 0, 说明去除评分列的na了
df.iloc[70:80]
```

### 15 - 缺失值补全｜整体均值填充

对于评价人数列的缺失值处理，我们可以使用整列的均值进行填充

-> 现在，将评价人数列的缺失值，用整列的均值进行填充

```python
na_idx_comment = list(df[df["评价人数"].isna()].index)
na_idx_comment
```

```python
df.loc[na_idx_comment, "评价人数"].fillna(value=df["评价人数"].mean(), inplace=False)
# df.iloc[na_idx_comment] # filled with col mean
```

### 16 - 缺失值补全｜上下均值填充

<br>

除了可以使用整列的均值进行填充，也可以使用缺失值位置的上下均值进行填充、

-> 现在，将评价人数列的缺失值，用上下数字的均值进行填充

```python
# df["评价人数"].interpolate()
df.loc[na_idx_comment, "评价人数"].fillna(value=df["评价人数"].interpolate(), inplace=False)
```
